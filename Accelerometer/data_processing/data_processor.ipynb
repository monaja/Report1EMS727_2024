```python
import matplotlib.pyplot as plt
import numpy as np

# Example data (angles in degrees, accelerometer z-axis values)
angles = [0, 24, 60, 72, 120, 144, 180, 204, 240, 276, 300, 336, 360]
z_values = [1.0000, 0.9135, 0.4848, 0.3090, -0.3090, -0.4848, -1.0000, -0.9135, -0.4848, -0.3090, 0.3090, 0.4848, 1.0000]
```

```python

# Add angles from 390° to 1080° with a 30° interval
angles_extended = angles + [390, 420, 450, 480, 510, 540, 570, 600, 630, 660, 690, 720]

# Manually calculate the Z-values using the cosine function for the additional angles
z_values_extended = z_values + [np.cos(np.radians(angle)) for angle in [390, 420, 450, 480, 510, 540, 570, 600, 630, 660, 690, 720]]
```

```python

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(angles_extended, z_values_extended, color='blue', label='Accelerometer Z-Axis')
```

```python

# Add labels and title
plt.title('Accelerometer Z-Axis Values at Different Angles', fontsize=14)
plt.xlabel('Angle (degrees)', fontsize=12)
plt.ylabel('Z-Axis Value', fontsize=12)

# Show grid
plt.grid(True)

# Show the plot
plt.legend()
plt.show()
```
